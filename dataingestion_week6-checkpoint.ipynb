```python
from pyaml import yaml

import dask.dataframe as dd
dde=dd.read_csv('en-fr.csv')
dde_to_dict=dde.compute()
dde_to_dict=dde_to_dict.to_dict(orient='records')
yaml_file=yaml.dump(dde_to_dict)
with open('output.yaml', 'w') as file:
    file.write(yaml_file)
```

```python
import pandas as pd
separator = ',' #delimiter if needed
dd=pd.read_csv('en-fr.csv')

col_names=list(dd.columns)
```

This next part is designed to import the yaml library, creates a dictionary with key columns and add it as a key to columns names.
it then dumps the column names into a yaml file named as output.yaml and saves the file.

```python
import yaml
dict_dat={'columns': col_names}
y_dat=yaml.dump(dict_dat)
with open('output.yaml', 'w') as file:
    file.write(y_dat)

```

To load file using yaml and compare column length and column names to vaidate the saving method, the following code was implemented. the code loads the yaml file presented in the directory and extracts the values saved under the columns key and save them into another list. the code uses the columns loaded from yaml and from pandas dataframe csv file to put into comparison.

```python
with open('output.yaml', 'r') as file:
    y_data = yaml.safe_load(file)
y_dat_col=y_data['columns']
if len(dd.columns) == len(y_dat_col):
    print("there is a match between column numbers")
for i in  range(0, len(dd.columns)):
    if col_names[i]==y_dat_col[i]:
        print(f'columns {i} match')
```
